```python
import inline as inline
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Draw
from matplotlib import pyplot as plt

file = pd.read_csv('ligands.csv.csv', index_col=[0])
# Importing a csv file from ligand from quadraduplex ligand database: https://www.g4ldb.com/ and extracting some data
table = file[['Molecular Weight','Polar Surface Area', 'HBA','HBD', 'Molecular Formula','Smiles','Bioactivities', 'Targets']]
table

```

```python
molecule_smiles = table['Smiles']
#print (molecule_smiles)
def make_smile(molecule_smiles):
    #Function which makes a 2D strcuture of chemical couponds from smiles str
    for molecule, indexx in zip(molecule_smiles, molecule_smiles.index) :
        #if smile != "" :
        print('\033[1m' +'Ligand: '+'\033[0m'+ indexx + '\033[1m'+'\nSmiles: '+'\033[0m'+ molecule , '\033[1m'+"\n2D structure:"+'\033[0m')
        img = Draw.MolToImage( Chem.MolFromSmiles(molecule), size=(200,200) )
        display(img)
    return display(img)

make_smile(molecule_smiles)
```

```python
import seaborn as sns
import numpy as np

table_corr=table[['Molecular Weight','Polar Surface Area', 'HBA','HBD']]
f, ax = plt.subplots(figsize=(10, 8))
corr = table_corr.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool),cmap="Greens",
            square=True, ax=ax, annot=True)
```
